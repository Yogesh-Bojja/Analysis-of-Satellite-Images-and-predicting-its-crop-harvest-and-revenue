```python
import geopandas as gpd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import shapefile
```

```python
gd = gpd.read_file("plots.shp")
```

```python
def show_plots(image, plot_plan_num):
    r = shapefile.Reader("plots.shp")
    gd = gpd.read_file("plots.shp")
    
    xdist = r.bbox[2] - r.bbox[0]
    ydist = r.bbox[3] - r.bbox[1]
    # Image width & height
    iwidth = 512
    iheight = 512
    xratio = iwidth/xdist
    yratio = iheight/ydist
    
    plot_list = []
    l = list(gd['plotsplan'])
    for i in range(len(l)):
        if l[i] == plot_plan_num:
            plot_list.append(i)

    if len(plot_list) == 0 :
        print("Invalid Plot-Plan number !!!")
        return 
    img = Image.new("RGBA", (iwidth, iheight), "white")
    draw = ImageDraw.Draw(img)

    for i in plot_list:
        pixels = []
        for x,y in r.shapes()[i].points:
            px = int(iwidth - ((r.bbox[2] - x) * xratio))
            py = int((r.bbox[3] - y) * yratio)
            pixels.append((px,py))
        draw.polygon(pixels, outline="rgba(240, 255, 0, 255)", 
                fill="rgba(255, 255, 255, 0)")
    img.save(".\plot_shapes\{}.png".format(plot_plan_num))
    bg = Image.open(image)
    ol = Image.open(".\plot_shapes\{}.png".format(plot_plan_num))
    ol_pixel = ol.load()

    for x in range(512):
        for y in range(512):
            if ol_pixel[x, y] == (255,255,255,255):
                ol_pixel[x, y] = (255,255,255,0)

    bg.paste(ol,(0,0), ol)
    bg.save(".\\final_images\\{}on{}.png".format(plot_plan_num, image[:-4]))
    print("Image saved!!!")
```

```python
show_plots("2016-12-22.png", "204-C12321")
```
